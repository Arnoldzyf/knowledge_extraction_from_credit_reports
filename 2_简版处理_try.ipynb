```python
import pdfplumber
import re
import sys
```

# Read All Text

- 入参
    - `pdf_file` 简版征信模板存储位置
    - `tail_page_num` 文档末尾有多少页是文件说明，不需要进行知识抽取。简版文档默认为1页。

```python
pdf_file = "./说明文件/简版征信模板.pdf"
tail_page_num = 1

text_all = ""

pdf_simple = pdfplumber.open(pdf_file)
page_range = range(len(pdf_simple.pages)-tail_page_num)
for i in page_range:
    page = pdf_simple.pages[i]
    text_all += page.extract_text() + '\n'

print(text_all)
```

```python
"""
删除`第n页/共m页`
"""
# 正则表达式匹配 "第n页/共m页" 的模式
page_pattern = r'第\d+页/共\d+页\n*'

# 使用 re.sub 替换所有匹配的模式为空字符串
text_all = re.sub(page_pattern, '', text_all)

print(text_all)
```

# 以`信用卡 发生过逾期的贷记卡账户明细如下：`为例

```python
sec_pattern = r"\n信用卡\n"

# 使用正则表达式查找所有 '\n信用卡\n' 的位置
matches = re.finditer(sec_pattern, text_all)
match_list = [match for match in matches]
if len(match_list) == 0:
    sys.exit("error: 无信用卡模块!")
if len(match_list) > 1:
    sys.exit("error: 有多个无信用卡模块!")
sec_start = match_list[0].end()
print(sec_start)
print(text_all[sec_start])
```

```python
sec_pattern = r"\n贷款\n"

# 使用正则表达式查找所有 '\n信用卡\n' 的位置
matches = re.finditer(sec_pattern, text_all)
match_list = [match for match in matches]
# print(match_list)
if len(match_list) == 0:
    sys.exit("error: 无信用卡模块!")
# if len(match_list) > 1:
#     sys.exit("error: 有多个无信用卡模块!")
sec_end = match_list[1].start()  ## 设置参数, 第几个为标题
print(sec_end)
print(text_all[sec_end])
```

```python
from paddlenlp import Taskflow
```

```python
schema = ['发卡日期', '发卡分行','账户币种','卡片尾号','销户日期']
schema.extend(['截至日期','信用额度','是否为呆账','余额','当前有无逾期'])
schema.extend(['近5年处于逾期状态月数','逾期超过90天月数'])
print(schema)              
              
few_ie = Taskflow('information_extraction', schema=schema, batch_size=16)

data_input="""
1. 2004 年 8 月 30 日中国建设银行北京分行发放的贷记卡（人民币账户，卡片尾号：0001）。截至 2010 年 10 月，信用额度\n10,000，已变成呆账，余额500。
"""
data_input = data_input.replace("\n","")
print(data_input)
results=few_ie(data_input)
print(results)

data_input="""
\n2. 2004年8月30日中国工商银行北京分行发放的贷记卡（人民币账户，卡片尾号：0002）。截至2016年7月，信用额度50,000，\n余额5,000（含未出单的大额专项分期余额4,000），当前有逾期。最近5年内有11个月处于逾期状态，其中5个月逾期超\n过90天。
"""
data_input = data_input.replace("\n","")
print(data_input)
results=few_ie(data_input)
print(results) ## 无法识别：是否为呆账

data_input="""
\n3. 2010年4月1日中国民生银行信用卡中心发放的贷记卡（人民币账户，卡片尾号：0003），2015年12月销户。最近5年内\n有7个月处于逾期状态，其中3个月逾期超过90天。"""
data_input = data_input.replace("\n","")
print(data_input)
results=few_ie(data_input)
print(results)
```
